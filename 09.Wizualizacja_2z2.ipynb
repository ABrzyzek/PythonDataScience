```python

import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(20,10))
# Plot circle or radius 3
plt.style.use("dark_background")

an = np.linspace(0, 2*np.pi, 100)

plt.subplot(221)
plt.plot(3*np.cos(an), 3*np.sin(an))
plt.title('nierówne osie, elipsa', fontsize=10)

plt.subplot(222)
plt.plot(3*np.cos(an), 3*np.sin(an))
plt.axis('equal')
plt.title('równe', fontsize=10)

plt.subplot(223)
plt.plot(3*np.cos(an), 3*np.sin(an))
plt.axis('equal')
plt.axis([-3, 3, -3, 3])
plt.title('okrąg, nawet po zmiane zakresów', fontsize=10)

plt.subplot(224)
plt.plot(3*np.cos(an), 3*np.sin(an))
plt.axis('equal')
plt.axis([-3, 3, -3, 3])
plt.plot([0, 4], [0, 2])
plt.title('równe, i to z linią', fontsize=10)

plt.show()
```

```python
df=miasta.set_index('Miasto').join(wspolrzedne.set_index('Miasto'))
```

```python
df.reset_index(inplace=True, drop=False)
```

```python
df.shape
```

```python
df
```

```python
def geo2float(s):
    s=s.replace('°','.')
    s=s[:5]
    degrees = int(s[:2])
    minutes = float(s)-degrees
    minutes *= (100/60)
    return float(degrees+minutes)
```

```python
geo2float("19°30'E")
```

```python
df['Szerokosc'] = df['Szerokosc'].apply(geo2float)
df['Dlugosc'] = df['Dlugosc'].apply(geo2float)
```

```python
df
```

```python
import numpy as np

df['s'] = df['ludnosc'].apply(np.sqrt)
df['s'] = (df['s'])/10
df
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt

proj = ccrs.Mercator()
fig = plt.figure(figsize=(20,10))
# ax = fig.add_subplot(1, 1, 1, projection=proj)
ax = plt.axes(projection=ccrs.Mercator())
plt.scatter(
    x=df["Dlugosc"],
    y=df["Szerokosc"],
    color="red",
    s=df['s'],
    alpha=1,
    transform=ccrs.PlateCarree()
)
extents = proj.transform_points(ccrs.Geodetic(),np.array([14, 24.5]),np.array([ 48.5, 55]) )

ax.set_extent([14, 24.5, 48.5, 55], crs=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle='-',  alpha=0.8)
ax.add_feature(cfeature.LAKES, alpha=0.5)
ax.add_feature(cfeature.RIVERS)


plt.show()

```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt

proj = ccrs.Mercator()
fig = plt.figure(figsize=(20,10))
# ax = fig.add_subplot(1, 1, 1, projection=proj)
ax = plt.axes(projection=ccrs.Mercator())
plt.scatter(
    x=df["Dlugosc"],
    y=df["Szerokosc"],
    color="red",
    s=df['s'],
    alpha=1,
    transform=ccrs.PlateCarree()
)
extents = proj.transform_points(ccrs.Geodetic(),np.array([14, 24.5]),np.array([ 48.5, 55]) )

ax.set_extent([14, 24.5, 48.5, 55], crs=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle='-',  alpha=0.8)
ax.add_feature(cfeature.LAKES, alpha=0.5)
ax.add_feature(cfeature.RIVERS)

miasta=df.nlargest(15, ['ludnosc'])
for row in  miasta.itertuples():
    ax.annotate(row.Miasto, xy=(row.Dlugosc, row.Szerokosc-0.05), xycoords=ccrs.PlateCarree()._as_mpl_transform(ax),
            ha='center', va='top', color='black')
plt.show()

```
