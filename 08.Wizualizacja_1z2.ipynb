```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(N) # the x locations for the groups
width = 0.35
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])

ax.bar(ind, menMeans, width, color='b')
ax.bar(ind, womenMeans, width,bottom=menMeans, color='r')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticklabels(['', 'G1', 'G2', 'G3', 'G4', 'G5'])
ax.set_yticks(np.arange(0, 81, 10))
ax.legend(labels=['Men', 'Women'])
plt.show();
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
a1 = fig.add_axes([0,0,2,1])
x = np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_ylabel('exp')
a2 = a1.twinx()
a2.plot(x, np.log(x),'ro-')
a2.set_ylabel('log')
fig.legend(labels = ('exp','log'),loc='upper left')
plt.show()
```

## Facetting

Facetting - rozdzielenie wykresu na kilka wykresów o tych samych osiach

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5)
fig = plt.figure()

fig.add_subplot(221)   # top left
plt.plot(np.cos(x))
fig.add_subplot(222)   # top right
plt.plot(np.sin(x))
fig.add_subplot(223)   # bottom left
plt.plot(np.tan(x))
fig.add_subplot(224)   # bottom right 
plt.plot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

day = pd.read_csv('.\data\day.csv')
```

```python
day
```

```python
plt.scatter('temp', 'cnt', data=day)
plt.xlabel('Znormalizowana Temperatura', fontsize='large')
plt.ylabel('Ilość wypożyczeń rowerów', fontsize='large');
```

```python
plt.scatter('temp', 'cnt', data=day, c='season')
plt.xlabel('Znormalizowana Temperatura', fontsize='large')
plt.ylabel('Ilość wypożyczeń rowerów', fontsize='large');
```

```python
plt.rcParams['figure.figsize'] = [15, 10]

fontdict={'fontsize': 18,
          'weight' : 'bold',
         'horizontalalignment': 'center'}

fontdictx={'fontsize': 18,
          'weight' : 'bold',
          'horizontalalignment': 'center'}

fontdicty={'fontsize': 16,
          'weight' : 'bold',
          'verticalalignment': 'baseline',
          'horizontalalignment': 'center'}

spring = plt.scatter('temp', 'cnt', data=day[day['season']==1], marker='o', color='blue')
summer = plt.scatter('temp', 'cnt', data=day[day['season']==2], marker='o', color='green')
autumn = plt.scatter('temp', 'cnt', data=day[day['season']==3], marker='o', color='orange')
winter = plt.scatter('temp', 'cnt', data=day[day['season']==4], marker='o', color='brown')

plt.legend(handles=(spring,summer,autumn,winter),
           labels=( 'Zima', 'Wiosna', 'Lato', 'Jesień'),
           title="Pora roku", title_fontsize=16,
           scatterpoints=1,
           bbox_to_anchor=(1, 0.7), loc=2, borderaxespad=1.,
           ncol=1,
           fontsize=14)

plt.title('Ilość wypożyczeń rowerów względem temperatury', fontdict=fontdict, color="white")
plt.xlabel("Znormalizowana temperatura", fontdict=fontdictx)
plt.ylabel("Ilość Wypozyczeń", fontdict=fontdicty);
```

### Ten wykres ma chyba następujące problemy
- dane się nakładają na siebie
- jest nieczytelny
- nie widać dobrze różnic między porami roku
- nie widać trendów
- wydaje się, że temperatura jest pozytywnie skorelowana z z ilością wypożyczeń

```python
fig = plt.figure()

plt.rcParams['figure.figsize'] = [15,10]
plt.rcParams["font.weight"] = "bold"

fontdict={'fontsize': 25,
          'weight' : 'bold'}

fontdicty={'fontsize': 18,
          'weight' : 'bold',
          'verticalalignment': 'baseline',
          'horizontalalignment': 'center'}

fontdictx={'fontsize': 18,
          'weight' : 'bold',
          'horizontalalignment': 'center'}

plt.subplots_adjust(wspace=0.2, hspace=0.2)

fig.suptitle('Ilość wypożyczeń rowerów względem temperatury\nz rozbiciem na pory roku', fontsize=25,fontweight="bold", color="white", 
             position=(0.5,1.01))

ax1 = fig.add_subplot(221)
ax1.scatter('temp', 'cnt', data=day[day['season']==1], c="blue")
ax1.set_title('Zima', fontdict=fontdict, color="blue")
ax1.set_ylabel("Ilość wypożyczeń", fontdict=fontdicty, position=(0,-0.1))


ax2 = fig.add_subplot(222)
ax2.scatter('temp', 'cnt', data=day[day['season']==2], c="green")
ax2.set_title("Wiosna", fontdict=fontdict, color="green")



ax3 = fig.add_subplot(223)
ax3.scatter('temp', 'cnt', data=day[day['season']==3], c="orange")
ax3.set_title('Lato', fontdict=fontdict, color="orange")


ax4 = fig.add_subplot(224)
ax4.scatter('temp', 'cnt', data=day[day['season']==4], c="brown")
ax4.set_title('Jesień', fontdict=fontdict, color="brown")
ax4.set_xlabel("Znormalizowana temperatura", fontdict=fontdictx, position=(-0.1,0));
```

## Tym razem
- Zimą jest pozytywna korelacja
- Ale na wiosnę i jesień jest lekka korelacja kwadratowa (?)
- W lecie ma widać jakichś trendów




```python
fig = plt.figure()

plt.rcParams['figure.figsize'] = [12,12]
plt.rcParams["font.weight"] = "bold"

plt.subplots_adjust(hspace=0.60)


fontdicty={'fontsize': 20,
          'weight' : 'bold',
          'verticalalignment': 'baseline',
          'horizontalalignment': 'center'}

fontdictx={'fontsize': 20,
          'weight' : 'bold',
          'horizontalalignment': 'center'}

fig.suptitle('Ilość wypożyczeń rowerów względem temperatury\nz rozbiciem na pory roku', fontsize=25,fontweight="bold", color="white", 
             position=(0.5,1.0))

#ax2 is defined first because the other plots are sharing its x-axis
ax2 = fig.add_subplot(412, sharex=ax2)
ax2.scatter('temp', 'cnt', data=day.loc[day['season']==2], c="green")
ax2.set_title('Wiosna', fontdict=fontdict, color="green")
ax2.set_ylabel("Ilość wypożyczeń", fontdict=fontdicty, position=(-0.3,-0.2))


ax1 = fig.add_subplot(411, sharex=ax2)
ax1.scatter('temp', 'cnt', data=day.loc[day['season']==1], c="blue")
ax1.set_title('Zima', fontdict=fontdict, color="blue")


ax3 = fig.add_subplot(413, sharex=ax2)
ax3.scatter('temp', 'cnt', data=day.loc[day['season']==3], c="orange")
ax3.set_title('Lato', fontdict=fontdict, color="orange")


ax4 = fig.add_subplot(414, sharex=ax2)
ax4.scatter('temp', 'cnt', data=day.loc[day['season']==4], c="brown")
ax4.set_title('Jesień', fontdict=fontdict, color="brown")
ax4.set_xlabel("Znormalizowana temperatura", fontdict=fontdictx);
```

- Zima ma najniższe temperatury!
- Lato ma najwyższe temperatury
- Wiosną i latem relacja jest kwadratowa (parabola)
- W niższych temperaturach - zawsze jest mniej pożyczanych rowerów
- Zimą jest wyraźna pozytywna korelacja 
- Latem mamy lekko negatywną korelację

```python
fig = plt.figure()

plt.rcParams['figure.figsize'] = [10,10]
plt.rcParams["font.weight"] = "bold"

plt.subplots_adjust(wspace=0.5)
fontdicty1={'fontsize': 18,
          'weight' : 'bold'}

fontdictx1={'fontsize': 18,
          'weight' : 'bold',
          'horizontalalignment': 'center'}

fig.suptitle('Ilość wypożyczeń rowerów względem temperatury\nz rozbiciem na pory roku', fontsize=25,fontweight="bold", color="white", 
             position=(0.5,1.0))

ax3 = fig.add_subplot(143, sharey=ax3)
ax3.scatter('temp', 'cnt', data=day.loc[day['season']==3], c="orange")
ax3.set_title('Lato', fontdict=fontdict,color="orange")


ax1 = fig.add_subplot(141, sharey=ax3)
ax1.scatter('temp', 'cnt', data=day.loc[day['season']==1], c="blue")
ax1.set_title('Zima', fontdict=fontdict, color="blue")
ax1.set_ylabel("Wypożyczenia rowerów", fontdict=fontdicty1, position=(0.5,0.5))


ax2 = fig.add_subplot(142, sharey=ax3)
ax2.scatter('temp', 'cnt', data=day.loc[day['season']==2], c="green")
ax2.set_title('Wiosna', fontdict=fontdict, color="green")


ax4 = fig.add_subplot(144, sharey=ax3)
ax4.scatter('temp', 'cnt', data=day.loc[day['season']==4], c="brown")
ax4.set_title('Jesień', fontdict=fontdict, color="brown")
ax4.set_xlabel("Znormalizowana temperatura", fontdict=fontdictx, position=(-1.5,0));
```

- W każdej porze roku dochodzi się do 8000 wypożyczeń
- Zimą i latem na wyraźnie clustry
- Wiosną i jesienią jest największy rozstrzał
